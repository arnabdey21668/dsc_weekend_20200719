```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = x *2 - 10
print(x)
print(y)

plt.plot(x, y, 'r', marker='*' )
plt.plot(x, y **2 , 'r', marker='*' )
plt.show()

plt.plot(x, y **3, marker='*' )
plt.show()

plt.plot(x, y, marker='*' )
plt.plot(x, y **2 ,marker='*' )
plt.show()
```

```python
x = np.arange(10)
y = x *2 - 10
print(x)
print(y)

plt.plot(x, y **2 , marker='*' )
plt.xticks(x)
plt.xlabel("Time in s")
plt.title("Graph")

plt.show()
```

```python
x = np.arange(10)
y = x *2 - 10
print(x)
print(y)

plt.plot(x, y **2 , linestyle='--' )

plt.plot(x, y **3 , 'go' )
plt.xticks(x)
plt.xlabel("Time in s")
plt.title("Graph")

plt.savefig("graph.png")
plt.show()

```

```python
plt.figure(figsize=(10,3))
x = np.arange(10)
y = x *2 - 10

# 2x3
# 1 2 3
# 4 5 6

# 1x2
# 1 2
plt.subplot(1,2, 1)
plt.plot(x,y)

plt.subplot(1,2, 2)
plt.plot(x,y**2)
plt.plot(x,y**3)

plt.show()
```

```python
#  legend
plt.figure(figsize=(10,3))
x = np.arange(10)
y = x *2 - 10

# 2x3
# 1 2 3
# 4 5 6

plt.plot(x,y, label = 'line-1')
plt.plot(x,y**2, label='line-2')

for i,j in zip(x,y):
    #plt.text(i,j, f"({i},{j})")
    #plt.text(i,j, "({},{})".format(i,j))
    plt.text(i,j+4, "(%d,%d)" % (i,j))

#plt.axhline()

plt.legend()
plt.show()
```

```python
x = list("ABCD")
y1 = pd.Series([10,5,50,4])
y2 = pd.Series([10,15,5,14])
y3 = pd.Series([4,25,15,40])

plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.bar(x, y1)

plt.subplot(2,2,2)
plt.bar(x, y2)

# stacked bar chart
plt.subplot(2,2,3)
plt.bar(x, y1)
plt.bar(x, y2, bottom= y1)
plt.bar(x, y3, bottom= y1 + y2)


# stacked using pandas dataframe
df = pd.DataFrame({"c1":y1, "c2":y2, "c3": y3})
print(df)
df.plot(kind='bar', stacked =True)

plt.show()

# adjacent bars
df = pd.DataFrame({"c1":y1, "c2":y2, "c3": y3})
print(df)
df.plot(kind='bar', width=0.9)

plt.show()
```

```python
x = np.linspace(1,10, 100)
y = x**2 - x*2 +20 
y1 = x**1.5
print(y)
plt.fill_between(x,y, alpha=0.8)
plt.fill_between(x,y + y1, y, alpha=0.8)
plt.show()
```
