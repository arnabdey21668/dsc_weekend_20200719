
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_plotting = pd.read_csv('titanic_filtered.csv')
```

```python
# plot histogram of ages of people

plt.figure(figsize=(15,6))
age1 = titanic_plotting.age
res = plt.hist(age1)

plt.show()
```

```python
# plot no of people in each pclass on bar chart

pclass_val = titanic_plotting['pclass'].value_counts()
pclass_val.plot(kind='bar')
```

```python
# plot a bar-chart depicting the ratio of males to females

gender_count = titanic_plotting['gender'].value_counts()
gender_count.plot(kind='bar')
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.

groupGender = titanic_plotting.groupby(by = ["gender"])
df = groupGender["survived"].value_counts()
maleSurvived= df.male[1]
maleNotSurvived= df.male[0]
femaleSurvived= df.female[1]
femaleNotSurvived= df.female[0]

x= pd.Series(["male-0","male-1","female-0","female-1"])
passangers = pd.Series([maleSurvived,maleNotSurvived,femaleSurvived,femaleNotSurvived])
plt.bar(x, passangers)
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.

groupGender = titanic_plotting.groupby(by = ["gender"])
df = groupGender["survived"].value_counts()
x= pd.Series(["male","female"])
passangerSurvived = pd.Series([df.male[1],df.female[1]])
passangerNotSurvived = pd.Series([df.male[0],df.female[0]])

plt.bar(x, passangerSurvived)
plt.bar(x, passangerNotSurvived, bottom= passangerSurvived)
```

```python
# plot stacked bar of survived with different pclass in same bar

groupGender = titanic_plotting.groupby(by = ["survived"])
df = groupGender["pclass"].value_counts()
survived = pd.Series([df[1][1],df[1][2],df[1][3]])
notSurvived = pd.Series([df[0][1],df[0][2],df[0][3]])

pclass = pd.Series(["1","2","3"])
plt.bar(pclass, survived)
plt.bar(pclass,notSurvived, bottom= survived)
```

```python
# plot box plot of ages of survived vs not-survived

plt.figure(figsize=(5,4))

a = titanic_plotting[titanic_plotting.survived == 1].age
print(a)


plt.boxplot(a)
plt.show()

b = titanic_plotting[titanic_plotting.survived == 0].age
print(b)


plt.boxplot(b)
plt.show()

plt.boxplot([a,b])
plt.show()
```

```python
# plot scatter plot of ages vs survived.

plt.figure(figsize=(25,6))
plt.scatter(titanic_plotting.age, titanic_plotting.survived)
plt.show()
```
