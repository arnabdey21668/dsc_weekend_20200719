```python

print("********"*10)
print("\n")
print("Q1.create a linearly spaced matrix M1 of size 4x4 having values in range 1-16")
M1= np.arange(1,17).reshape(4,4)
print(M1)

print("********"*10)
print("\n")
print("Q2.create a transpose of above matrix call it M2")
M2=M1.transpose()
print(M2)

print("********"*10)
print("\n")
print("Q3.find sum of above matrix M3 = (M1 + M2")

M3=M1+M2
print(M3)

print("********"*10)
print("\n")
print("Q4.Find Transpose of M3, Call it MT1. Check if M3 == MT1")
MT1=M3.transpose()

if (M3==MT1).all()==True:
    print("M3,MT1 are matching")
else:
    print("M3,MT1 are not matching")
# count=0
# for i in range(M3.itemsize):
#     for j in range(MT1.itemsize):
#         if M3[i][j]==MT1[i][j]:
#             count+=1

# if count==MT1.size:
#     print("It's a match")
# else:
#     print("Not matching")


print("********"*10)
print("\n")
print("Q5.# find diffrence of M4 = (M1 - M2)")

M4=M1-M2
print(M4)

print("********"*10)
print("\n")
print("Q6.# Find Transpose of M4, Call it MT2. Check if M4 == MT2. Also check if M4 == -MT2")

MT2=M4.transpose()
print('MT2\n',MT2)

#Method1
if (M4==MT2).all()==True:
    print("M4,MT2 are matching")
else:
    print("M4,MT2 are not matching")
    
#Method2
if np.array_equal(M4,MT2)==True:
    print("M4,MT2 are matching")
else:
    print("M4,MT2 are not matching")

if np.array_equal(M4,-MT2)==True:
    print("M4 is negation of MT2")
else:
    print("M4 is Not negation of MT2")

```
