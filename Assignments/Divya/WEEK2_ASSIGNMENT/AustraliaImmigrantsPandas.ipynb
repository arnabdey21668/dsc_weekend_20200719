```python
import numpy as np
import pandas as pd
```

```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows.
# List all Column and Row indexes(Print their names)

aus_df = pd.read_excel('C:/Users/DIVYA/Documents/DataScienceCourse/Python/numpy/dsc_weekend_20200719/Assignments/Divya/WEEK2_ASSIGNMENT/Australia.xlsx',sheet_name='Australia by Residence',skiprows=20)
print(aus_df.columns)
print(list(aus_df.index))
```

Dataset Analysis

```python
# Describe the dataset

aus_df.describe()
```

```python
# Use values and index attributes on both Row and Column Indexes
print(aus_df.values)
print()
print(aus_df.index)
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

print(aus_df.shape)
print()
print(aus_df.dtypes)
```

Drop



```python
# - Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
filter = ((aus_df['Type']=='Emigrants' ) & (aus_df['Type']  is not 'Immigrants') )
aus_df.drop(aus_df[filter].index,inplace=True)
print(aus_df)
print()
print(aus_df.size)
```

# Find columns for which more than 50% data is missing and drop them (Column Deletion)
mask = aus_df.isna()
(mask.any()).sum()


```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
print("Dataframes before deletion \n\n",aus_df.head(5))
aus_df.drop(['Type','Coverage','AREA','REG','DEV'],axis=1,inplace=True)
print()
print("Dataframes after deletion \n\n",aus_df.head(5))
```

## Rename

```python
#Rename columns :OdName -> Country; RegName -> Region
print("Dataframes before rename \n\n",aus_df.head(5))
print()
aus_df.rename(columns={'OdName' : 'Contry','RegName':'Region'},inplace=True)
print("Dataframes after  rename \n\n",aus_df.head(5))
```

```python
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
a = list((aus_df .iloc[:, 4 :39]).columns)
b = list(map(str, a)) # => [1,2,3]
d = dict(zip(a,b))
aus_df.rename(columns=(d),inplace=True)
print(aus_df.columns)
```

```python
# - Find Total number of immigrants for all years, and create a new column Total in data Frame
aus_df['Total'] = aus_df[b].sum(axis=1)

aus_df['Total']
aus_df
```

```python
# Sort Data by the new column Total in descending order and analyze the fist few rows.
aus_df.sort_values('Total',ascending=False).head()
# find something weird in the dataset ? Delete that row 'world'

mask =  aus_df['AreaName']=='World'

aus_df.drop(aus_df[mask].index,inplace=True)

#Analyze last few rows also in the sorted result (Don't sort inplace)
aus_df.sort_values('Total',ascending=False).tail()
```
