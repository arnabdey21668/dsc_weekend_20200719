
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data generated previously from 'titanic_filtered.csv'

df = pd.read_csv('titanic_filtered.csv')
```

```python
# plot histogram of ages of people

df.age.plot(kind="hist", bins=20)
plt.xlabel('Age')
plt.ylabel('Count of people')
plt.show()
```

```python
# plot no of people in each pclass on bar chart

df.pclass.value_counts().plot(kind="bar")
plt.xlabel('pclass')
plt.ylabel('Count of people')
plt.show()
```

```python
# plot a bar-chart depicting the ratio of males to females

df.gender.value_counts().plot(kind="bar")
plt.ylabel('Count of people')
plt.show()
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.

grp = df.groupby(by=["gender","survived"])
grp.size().plot(kind="bar")
plt.ylabel('Count of people')
plt.show()
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.

df_new = df.groupby(by=['gender'])['survived'].value_counts()
x= pd.Series(["male","female"])

Alive = pd.Series([df_new.male[1], df_new.female[1]])
Dead = pd.Series([df_new.male[0], df_new.female[0]])

plt.bar(x, Alive, label='Alive')
plt.bar(x, Dead, bottom= Alive, label='Dead')
plt.ylabel('Count of people')
plt.legend()
plt.show()
```

```python
# plot stacked bar of survived with different pclass in same bar

df_new = df.groupby(by = ["survived"])["pclass"].value_counts()

x= pd.Series(["1","2","3"])
Alive = pd.Series([df_new[1][1],df_new[1][2],df_new[1][3]])
Dead = pd.Series([df_new[0][1],df_new[0][2],df_new[0][3]])

plt.bar(x, Alive, label='Alive')
plt.bar(x, Dead, bottom= Alive, label='Dead')
plt.xlabel('pclass')
plt.ylabel('Count of People')
plt.legend()
plt.show()
```

```python
# plot box plot of ages of survived vs not-survived

Alive = df[df['survived']==1].age
Dead = df[df['survived']==0].age
plt.boxplot([Alive, Dead])
plt.ylabel('Age')
plt.show()
```

```python
# plot scatter plot of ages vs survived.

plt.figure(figsize=(20,3))
plt.scatter(df['age'],df['survived'])
plt.xlabel('Age')
plt.ylabel('Sruvived')
plt.show()
```
